```python
import pandas as pd
```

```python
data = pd.read_csv("organize_hexa.csv",sep=',', header=None)
```

```python
data
```

```python
data.columns = ['species','mean','std']
data.head(21)
```

```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.pyplot import figure
import matplotlib
fig = matplotlib.pyplot.gcf()
figure(figsize=(20, 15), dpi=80)
fig.set_size_inches(50, 30, forward=True)
x = range(len(data['species']))
plt.errorbar(x, data['mean'], yerr = data['std'],fmt='sk',
              elinewidth=1,lw=10, capsize=0)
plt.xticks(x, data['species'],fontsize=14, rotation=90)
plt.yticks(fontsize=14)

plt.grid()
N=150
plt.gca().margins(x=0.01)
plt.gcf().canvas.draw()
tl = plt.gca().get_xticklabels()
maxsize = max([t.get_window_extent().width for t in tl])
m = 0.2 # inch margin
s = maxsize/plt.gcf().dpi*N+2*m
margin = m/plt.gcf().get_size_inches()[0]

#plt.gcf().subplots_adjust(left=margin, right=1.-margin)
plt.gcf().set_size_inches(s, plt.gcf().get_size_inches()[1])

fig.subplots_adjust(bottom=0.7)
plt.title('Mean and Standard Deviation for the Length of Hexanucleotides', fontdict = {'fontsize' : 40})
plt.savefig("Hexanucleotide.png",dpi=300)
plt.show()
```
